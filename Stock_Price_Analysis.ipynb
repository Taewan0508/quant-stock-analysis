```python
#import all libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import os

#Download apple stock data during the time range and returns pandas DataFrame
data = yf.download("AAPL", start="2020-01-01", end="2024-12-31")

#Calculate daily return by computing percentage change between current and previous closing price
data["Daily Return"] = data['Close'].pct_change()

#Calculate the 20-day moving average (average closing price of last 20-days)
data["MA20"] = data['Close'].rolling(window=20).mean()

#Plot the data (only closing price of moving average 20-days)
data[['Close', 'MA20']].plot(figsize=(12,6))

#Create a folder for saving results
os.makedirs("results", exist_ok=True)

#Save the chart as an image
plt.savefig("results/aapl_moving_average.png")
```
